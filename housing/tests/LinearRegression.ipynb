```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([[1, 0.5], [2, 2], [3, 4.5], [4,8]])
y = np.dot(X, np.array([3, 4.5])) + 3

lin_reg = LinearRegression()
```

```python
lin_reg.fit(X, y)

lin_reg.intercept_
```

```python
lin_reg.coef_
```

```python
lin_reg.predict([[0.5, 1]])
```
