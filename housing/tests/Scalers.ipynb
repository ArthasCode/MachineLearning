## imported libraries for the test and initial data frame

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd

data = pd.DataFrame({
    "Age": [12, 18, 26, 56],
    "Name": ["João", "Marcelo", "Pedro", "Pablo"]
})

data_nums = data.drop(columns="Name")
```

# Min-Max Scaling

```python
scaler1 = MinMaxScaler(feature_range=(0, 2))
scaler2 = MinMaxScaler()

# É necessário retirar as colunas strings,
# pois o estimador ainda pede um array bidimensional.

data_scaled1 = scaler1.fit_transform(data_nums)
data_scaled2 = scaler2.fit_transform(data_nums)

print(data_scaled1)
print(data_scaled2)
```

## Parameters learned with the provided data

```python
print(scaler1.min_)
print(scaler2.min_)
```

```python
print(scaler1.scale_)
print(scaler2.scale_)
```

```python
print(scaler1.data_max_)
print(scaler2.data_max_)
```

# Standard Scaling

```python
scaler1 = StandardScaler()
scaler2 = StandardScaler(with_mean=False)
scaler3 = StandardScaler()

data_scaled1 = scaler1.fit_transform(data_nums)
data_scaled2 = scaler2.fit_transform(data_nums)

print(data_scaled1)
print(data_scaled2)

# Exemplo com outliers
data_nums_outlier = data_nums.copy()
data_nums_outlier.loc[3, "Age"] = 120
data_scaled3 = scaler3.fit_transform(data_nums_outlier)

print(data_scaled3)
```

```python
print(scaler1.var_)
print(scaler2.var_)
print(scaler3.var_)
```

```python
print(scaler1.scale_)
print(scaler2.scale_)
print(scaler3.scale_)
```

```python
print(scaler1.mean_)
print(scaler2.mean_)
print(scaler3.mean_)
```
